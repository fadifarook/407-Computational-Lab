## Polynomial fitting using numpy

I will create a mock noisy dataset, then do simple linear and quadratic fits to it.

```python
import numpy as np

# mock dataset: relation between x and y is linear with noise
x = np.arange(50.)
y = x + 25.0 # * np.random.random(50)
```

```python
# plot what the mock dataset looks like
from matplotlib.pyplot import plot
plot(x, y, 'o')
```

```python
# Now let's try a linear fit
linfit = np.polyfit(x,y,1)
# linfit is array: first element slope, second element y-intercept
# so best-fit line is: y = linfit[0]*x + linfit[1]
linfit
```

```python
# Now let's try a quadratic fit
quadfit = np.polyfit(x,y,2)
# quadfit is array: quadratic coefficient, linear coefficient, constant
# so best-fit line is: y = linfit[0]*x^2 + linfit[1]*x + linfit[2]
quadfit
```

```python
# along the best-fit line or curve, we can find the y values corresponding to each of our x values
y_linfit = np.polyval(linfit, x)
y_linfit
```

```python
y_quadfit = np.polyval(quadfit, x)
y_quadfit
```

```python
# Now let's visualize the best-fit results
plot(x, y, 'o')
plot(x,y_linfit, y_quadfit)
```

As we would expect (because the mock dataset was ~ linear), the quadratic coefficient is very close to zero, so gives almost the identical result to the linear fit

* numpy.polynomial has similar functions to fit to: Chebyshev series, Hermite series, Laguerre series, Legendre series, etc.
